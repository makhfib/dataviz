```python
import numpy as np # linear algebra
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

from wordcloud import WordCloud, STOPWORDS

stopwords = set(STOPWORDS)
data = pd.read_csv("data/most_backed.csv")

wordcloud = WordCloud(
    background_color='white',
    stopwords=stopwords,
    max_words=200,
    max_font_size=40, 
    random_state=42
    ).generate(str(data['title']))


fig = plt.figure()
plt.imshow(wordcloud)
plt.axis('off')
fig.set_facecolor("white")
fig.savefig("figures/wordcloud.png", dpi=900, facecolor=fig.get_facecolor())
```
